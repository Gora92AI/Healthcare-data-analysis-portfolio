```python
import pandas as pd

data = pd.read_csv("C:/Users/diopg/OneDrive/Documents/Documents Health Projects/heart_disease_uci.csv")
print(data.columns)
```

```python
# 1. Visualization Sex (Gender) Distribuation
import matplotlib.pyplot as plt

data['sex'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Number of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()                                           
```

```python
# 2. Plot Age Distribution

data['age'].hist(bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# 3. Analyze Heart Disease Prevalence

print(data['num'].value_counts())
data['num'].value_counts().sort_index().plot(kind='bar', color='salmon')
plt.title('Heart Disease Diagnosis Distribution')
plt.xlabel('Diagnosis Code (0 = No Disease)')
plt.ylabel('Number of Patients')
plt.show()
```

```python
# CREATE A NEW COLUMN: 0 = No disease, 1 = Disease Present
data['disease_present'] = (data['num'] > 0).astype(int)
```

```python

# Count and Plot
data['disease_present'].value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title('Heart Disease Presence')
plt.xticks([0,1], labels=['No Disease', 'Disease'])
plt.ylabel('Number of Patients')
plt.show()
```

```python
n_with_disease = data['disease_present'].sum()
n_total = len(data)
pct_with_disease  = (n_with_disease / n_total) * 100

print(f"number of patient with heart disease: {n_with_disease}")
print(f"Total patients: {n_total}")
print(f"Percentage with heart disease: {pct_with_disease:.1f}%")
```

```python
# 4. Average Age by Disease Status

avg_ages = data.groupby('disease_present')['age'].mean()
print("Average age (no disease, disease): ", avg_ages.values)
```

```python
# 5. Gender Ratio by Disease

pd.crosstab(data['sex'], data['disease_present']).plot(kind='bar', stacked=True)
plt.title('Heart Disease by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.legend(['No Disease', 'Disease'])
plt.show()
```
